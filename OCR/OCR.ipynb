```python
import easyocr
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Read image

img = cv2.imread('school.jpg')
```

```python
# Shape of Image
img.shape
```

```python
# Show Image
cv2.imshow('School Board',img)
cv2.waitKey(0)
```

```python
# Build OCR


reader = easyocr.Reader(['en'])
result = reader.readtext('school.jpg')
result
```

```python
# Draw Bounding Boxes

top_left = tuple(result[0][0][0])
bottom_right = tuple(result[0][0][2])
text = result[0][1]

img = cv2.imread('school.jpg')
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, cv2.FONT_HERSHEY_SIMPLEX, 0.5,(255,255,255),2,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```

```python
# For multiple frames

reader = easyocr.Reader(['en'])
result = reader.readtext('sign board.jpg')
img = cv2.imread('sign board.jpg')
spacer = 150
for detection in result: 
    top_left = tuple(detection[0][0])
    bottom_right = tuple(detection[0][2])
    text = detection[1]
    img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),4)
    img = cv2.putText(img,text,(30,spacer), cv2.FONT_HERSHEY_SIMPLEX, 0.5,(0,255,0),3,cv2.LINE_AA)
    spacer+=15
    
plt.imshow(img)
plt.show()
```
